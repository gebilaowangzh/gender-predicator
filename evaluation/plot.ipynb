```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
```

# Plot logistic model

```python
df = pd.read_excel('logistic_regression.xlsx')
df.head()
```

```python
df = df[df['include undefined gender']=='Y']
df
```

```python
model1 = df[df['model']=='first name']
model2 = df[df['model']=='full name']
```

```python
x = model1[model1['input']=='full names']['dataset']
x
```

```python
y1_full = model1[model1['input']=='full names']['accuracy']
y1_first = model1[model1['input']=='first names']['accuracy']

y2_full = model2[model2['input']=='full names']['accuracy']
y2_first = model2[model2['input']=='first names']['accuracy']
```

```python
plt.plot(x, y1_full, 'dr')
plt.plot(x, y1_first, 'db')
plt.plot(x, y2_full, 'or')
plt.plot(x, y2_first, 'ob')

plt.scatter([], [], c='gray', marker='d', label='first name model')
plt.scatter([], [], c='gray', marker='o', label='full name model')
plt.scatter([], [], c='r', marker='.', label='input: full names')
plt.scatter([], [], c='b', marker='.', label='input: first names')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title('Accuracy of LR models trained on first/full names')
plt.xlabel('Dataset', fontsize=8)
plt.ylabel('Accuracy', fontsize=8)
plt.savefig('lr_model_comp.png', dpi=300, bbox_inches='tight')
```

# Plot Naive Bayes model

```python
df = pd.read_excel('naive_bayes.xlsx')
df = df[df['include undefined gender']=='Y']
df = df[df['method'] == 'lap']
df
```

```python
model1 = df[df['model']=='first name']
model2 = df[df['model']=='full name']
```

```python
x = model1[model1['dataset']=='train']['input']
x
```

```python
y1_train = model1[model1['dataset']=='train']['accuracy']
y1_test = model1[model1['dataset']=='test']['accuracy']

y2_train = model2[model2['dataset']=='train']['accuracy']
y2_test = model2[model2['dataset']=='test']['accuracy']
```

```python
plt.plot(x, y1_train, 'dr')
plt.plot(x, y1_test, 'db')
plt.plot(x, y2_train, 'or')
plt.plot(x, y2_test, 'ob')

plt.scatter([], [], c='gray', marker='d', label='first name model')
plt.scatter([], [], c='gray', marker='o', label='full name model')
plt.scatter([], [], c='r', marker='.', label='train')
plt.scatter([], [], c='b', marker='.', label='test')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title('Accuracy of bayes models trained on first/full names')
plt.xlabel('Dataset', fontsize=8)
plt.ylabel('Accuracy', fontsize=8)
plt.savefig('bayes_model_comp.png', dpi=300, bbox_inches='tight')
```
