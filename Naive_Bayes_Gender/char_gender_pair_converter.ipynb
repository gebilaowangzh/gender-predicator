```python
from collections import defaultdict, Counter
import json
```

```python
# combine train & dev sets as train set since dev set is not needed 
# and make a dict that pairs char with gender


genderDict = defaultdict(Counter)


def load_char(file):
    '''saved in this format: {'char': {'M': count1, 'F': count2, 'U': count3}}.
    
    Paras:
        file: filepath to the file you want to load
    '''
    global genderDict
    
    # pair char with gender
    def load(char, gender):
        for c in char:
            genderDict[c][gender] += 1
            # count the total num of char for each gender
            genderDict['total'][gender] += 1
    
    f = open(file, 'r')
    next(f)
    for line in f:
        line = line.split('\t')
        if line[3].strip() == 'M':
            load(line[1], 'M')
        elif line[3].strip() == 'F':
            load(line[1], 'F')
        else:
            load(line[1], 'U')

    return genderDict
```

```python
load_char('train_dev_test/train_ds.txt')
load_char('train_dev_test/dev_ds.txt')
```

```python
# save the dict as a json file so that the data can be reused easily

with open('dict4Gender.json', 'w') as f:
    json.dump(genderDict, f)
```
